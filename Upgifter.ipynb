```python
#5
import numpy as np

f1 = lambda z: 1.2 * z + np.sqrt(z) - 2
space = np.linspace(0, 2)
y1 = f1(space)
import scipy.optimize as opt

ns = opt.fsolve(f1, 0)
import matplotlib.pyplot as plt

plt.axhline(0)
plt.plot(space, y1, "g")
plt.plot(ns, 0, "r o")  # visa nollstllet
plt.grid()
plt.xlabel("X-Axeln")
plt.ylabel("Y-Axeln")
```

```python
#6
import numpy as np

f1 = lambda z: np.e ** -z * np.sin(z)
space = np.linspace(0, 6)
y1 = f1(space)
import scipy.optimize as opt

mpx = opt.minimize(f1, 4).x  #hitta minimipunkten
mpy = f1(mpx)  #Hitta Y vrdet fr minimipunkten
import matplotlib.pyplot as plt

plt.axhline(0)
plt.plot(space, y1, "g")
plt.plot(mpx, mpy, "r o")  # visa minimipunkten
plt.grid()
plt.xlabel("X-Axeln")
plt.ylabel("Y-Axeln")
```
